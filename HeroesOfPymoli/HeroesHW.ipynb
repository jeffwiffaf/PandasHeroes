```python
age_group_counts = age_group["Age Group"].value_counts()

age_group_counts_df = pd.DataFrame(age_group_counts)

age_group_counts_df = age_group_counts_df.rename(columns={"Age Group":"Counts"})

age_group_counts_df = age_group_counts_df.reset_index(level=["Age Group"])
age_group_counts_df = pd.DataFrame(age_group_counts_df["Age Group"].value_counts())
age_group_counts_df = age_group_counts_df.rename(columns={"Age Group":"Total Count"})
age_group_counts_df = age_group_counts_df.reindex(["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"])

#adding a column to the dataframe age_group_counts_df called "Percentage of Players"
# and will figure out the formula for you.

age_group_counts_df["Percentage of Players"] = (age_group_counts_df["Total Count"]/age_group_counts_df["Total Count"].sum())

#Formatting the value in the summary table
age_group_counts_df['Percentage of Players'] = pd.to_numeric(age_group_counts_df['Percentage of Players'])
age_group_counts_df['Percentage of Players'] = age_group_counts_df['Percentage of Players'].map("{:.2%}".format)

#age_group_counts_df
```
